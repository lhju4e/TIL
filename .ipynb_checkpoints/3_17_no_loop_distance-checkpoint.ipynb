### 1.  

~~~
    def compute_distances_no_loops(self, X):

    dists = -2 * np.dot(X, self.X_train.T) + np.sum(self.X_train **2,axis=1) + np.sum(X**2, axis=1)[:, np.newaxis]
    
    return dists
~~~
    
>  https://medium.com/dataholiks-distillery/l2-distance-matrix-vectorization-trick-26aa3247ac6c


\\({(x-y)}^{2} = {x}^{2} + {y}^{2} - 2xy \\)

출저를 보면, 곱셈공식을 이용해서 위의 함수를 작성한다고 한다. 

잠깐 곱셈공식이 떠올랐지만 행렬의 연산에서는 당연히 곱셈공식이 적용되지 않는다고 생각해서 풀지 못했는데, 이처럼 행렬 요소 연산(요소끼리의 거리 계산)인 경우에는 될 수도 있나보다.,

-------------------------------------
`
-2 * np.dot(X, self.X_train.T)   => -2xy 
`

요소끼리의 곱셈이므로, train데이터를 전치시켜서 곱한다. 그 결과 (500,5000)이 된다.

---------------------------------------
`
np.sum(self.X_train **2, axis=1) =>` \\({y}^{2}\\)


train행렬의 요소들을 제곱한 뒤, axis=1로 더한다. 그 결과, (5000,)이 된다.

---------------------------------------------------

+

axis = 0 : 가장 밖 괄호를 없앤다. 3차원에서는 x축을 없앤다고 생각하면 된다. (3, 2)행렬이 존재할 때, 같은 열끼리 더해서 (3,)으로 만들어준다.

axis = 1 : 밖에서 두 번째 괄호를 없앤다. 3차원에서는 y축을 없앤다고 생각하면 된다. (3,2)행렬이 존재할 때, 같은 행끼리 더해 (2,)으로 만들어준다.

axis = 2 : 밖에서 세 번째 괄호를 없앤다. 3차원에서의 z축을 없앤다고 생각하면 된다.

------------------------------------------------------


sum()으로 인해 (3,)이 된 행렬은, 차원이 하나 깎인(?) 상태. 후에 broadcast로 (1, 3)이 될 수도 있고, reshape 등으로 (3, 1)이 될 수도 있다. 


------------------------------------------------

`
np.sum(X**2, axis=1)[:, np.newaxis] => ` \\({x}^{2}\\)

test행렬들의 요소를 제곱한 뒤, axis=1로 더한다. 행끼리 더하므로, 그 결과 (500,)이 된다.

여기서, `[:, np.newaxis]`를 추가로 해 주는데, 차원을 맞춰주기 위함이다. 이 결과, (500, 1)이 된다.

(500,)는 (1, 500)이랑 같게 취급되는 듯, (500, x)랑 덧셈이 성립하지 않는다. 따라서 (500, 1)로 바꿔주어야 한다.

-------------------------------------------

앞에서부터 더하면 (500, 5000) + (5000, ) => (500, 5000) + (1, 5000)으로 브로드캐스팅 되어, 덧셈이 된다.

(500, 5000) + (500, 1) => 마찬가지로 브로드캐스팅 되어, (500, 5000)이 된다.

--------------------------------------------------------------------

+

np.newaxis는 행이나 열에 축?을 추가해 준다고 생각하면 되는 것 같다. 차원이 늘어난다.

ex)
sum 등으로 인해 (2,)가 되었을 때 [:,np.newaxis]를 하면 (2, 1)이 된다. 
+ 참고로, (2,)인 행렬과 (2,2)행렬을 덧샘하면 (1,2)로 브로드캐스팅 된 후 (2,2)로 더해진다.

그냥 1차원 리스트인 [1,2,3]에 [np.newaxis]를 하면 (1, 3)이 되고, [:,np.newaxis]를 하면 (3, 1)이 된다.(차원 증가)

(2,2)행렬에 [np.newaixs]를 하면 (1, 2, 2)가, [:, np.newaxis]를 하면 (2, 1, 2)가 된다.

### 2.

과제를 하다가 이상한 부분을 발견했다.

for문을 적게 쓸수록, 시간이 빨라져야 하는데 two loop가 one loop보다 빠르다.

non - loop는 압도적으로 빠르지만, 이상해서 열심히 찾아보았다.

다른 사람들이 질문을 많이 해놓았다. 그 중 하나의 답변이다. (주소 날아감;)

The reason is size of arrays within the loop body. In the two loop variant works on two arrays of 3000 elements. This easily fits into at least the level 2 cache of a cpu which is much faster than the main memory but it is also large enough that computing the distance is much slower than the python loop iteration overhead.

The second case the loop body works on 5000 * 3000 elements. This is so much that the data needs go to main memory in each computation step (first the Y-X[i] subtraction into a temporary array, squaring the temporary into another temporary and then read it back to sum it). The main memory is much slower than the cpu for the simple operations involved so it takes much longer despite removing a loop. You could speed it up a bit by using inplace operations writing into preallocated temporary array, but it will still be slower than the two loop variant for these array sizes.

요약하자면, 행렬이 너무 커서 for문으로 나눠서 계산하지 않고 한 번에 계산하려면 메인메모리로 보내야 하는데, 메인메모리는 캐시보다 느리니까, for문을 줄였어도 오히려 더 느려지는 현상이 나타날 수 있다는 것 같다.. 신기하다
